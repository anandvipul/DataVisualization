## Basic Plotting: plot

```python
ts = pd.Series(np.random.randn(1000),
              index = pd.date_range('1/1/2000', periods = 1000))
```

```python
ts = ts.cumsum()
```

```python
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000,4),
                 index = ts.index, columns = list('ABCD'))
```

```python
df = df.cumsum()
```

```python
plt.figure();
```

```python
df.plot();
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), 
                   columns = ['B', 'C']).cumsum()
```

```python
df3['A'] = pd.Series(list(range(len(df))))
```

```python
df3.plot(x = 'A', y = 'B')
```

## Other Plots

other plots available in pd.plot command with kind = '...'
- 'bar' or 'barh' for bar plots
- ‘hist’ for histogram
- ‘box’ for boxplot
- ‘kde’ or ‘density’ for density plots
- ‘area’ for area plots
- ‘scatter’ for scatter plots
- ‘hexbin’ for hexagonal bin plots
- ‘pie’ for pie plots

```python
plt.figure();
df.iloc[5].plot(kind = 'bar');
```

```python
df.iloc[5]
```

there are several plotting functions in pandas.plotting that take a Series or DataFrame as an argument. These include:

-     Scatter Matrix
-     Andrews Curves
-     Parallel Coordinates
-     Lag Plot
-     Autocorrelation Plot
-     Bootstrap Plot
-     RadViz

## Scatter Plot

```python
df = pd.DataFrame(np.random.rand(50, 4), \
                  columns = ['a', 'b', 'c', 'd'])
```

```python
df.plot.scatter(x = 'a', y = 'b');
```

```python
ax = df.plot.scatter(x = 'a', y = 'b', \
                     color = 'DarkBlue', label = 'Group 1');
df.plot.scatter(x = 'c', y = 'd', \
                color = 'DarkGreen', label = 'Group 2', \
                ax = ax)
```

```python
df.plot.scatter(x = 'a', y = 'b', c = 'c', s = 10);
```

```python
df.plot.scatter(x = 'a', y = 'b', s = df['c']*100);
```

## Hexagonal Bin Plot

```python
df = pd.DataFrame(np.random.randn(1000, 2), \
                  columns = ['a', 'b'])
```

```python
df['b'] = df['b'] + np.arange(1000)
```

```python
df.plot.hexbin(x = 'a', y = 'b', gridsize = 20);
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), \
                 columns = ['a', 'b'])
```

```python
df['b'] = df['b'] + np.arange(1000)
```

```python
df['z'] = np.random.uniform(0, 3, 1000)
```

```python
df.plot.hexbin(x = 'a', y = 'b', C = 'z', \
               reduce_C_function = np.max, \
               gridsize=25)
```
